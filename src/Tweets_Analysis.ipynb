```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk import FreqDist
from wordcloud import WordCloud
import matplotlib.pyplot as plt

df = pd.read_csv("../data/Corona_NLP_test.csv")
text_corpus = df["OriginalTweet"].str.lower().str.cat(sep=" ")

tokens = nltk.word_tokenize(text_corpus)
print(tokens[0:50])
```

```python
stop_words = set(stopwords.words("english"))

filtered_tokens = [token for token in tokens if token not in stop_words]
print(filtered_tokens[0:50])
```

```python
fdist = FreqDist(filtered_tokens)

# Print the 10 most common words
print(fdist.most_common(10))
```

```python
# Create a word cloud object
wordcloud = WordCloud(width=800, height=800, background_color="white").generate_from_frequencies(fdist)

# Plot the word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```
